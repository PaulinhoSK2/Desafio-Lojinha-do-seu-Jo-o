# Task
Analyze the revenue data from the files "df_2020.xlsx", "df_2021.xlsx", "df_2022.xlsx", and "df_2023.xlsx" by combining the data, calculating the total revenue per date, and plotting the total revenue over time.

## Combine dataframes

### Subtask:
Combine the four dataframes into a single dataframe.


**Reasoning**:
Concatenate the four dataframes into a single dataframe and display the head.



```python
df_list = [loja, loja2, loja3, loja4]
df_combined = pd.concat(df_list)
display(df_combined.head())
```

## Convert date column

### Subtask:
Convert the 'Data da Compra' column to datetime format.


**Reasoning**:
Convert the 'Data da Compra' column to datetime format in the combined DataFrame.



```python
df_combined['Data da Compra'] = pd.to_datetime(df_combined['Data da Compra'], format='%d/%m/%Y')
display(df_combined.info())
```

## Calculate total revenue

### Subtask:
Calculate the total revenue for each row by multiplying 'Preço' and 'Quantidade'.


**Reasoning**:
Create a new column 'Total Revenue' by multiplying 'Preço' and 'Quantidade de parcelas' and display the head of the dataframe.



```python
df_combined['Total Revenue'] = df_combined['Preço'] * df_combined['Quantidade de parcelas']
display(df_combined.head())
```

## Group by date

### Subtask:
Group the data by date and sum the total revenue.


**Reasoning**:
Group the dataframe by date and sum the total revenue.



```python
daily_revenue = df_combined.groupby('Data da Compra')['Total Revenue'].sum().reset_index()
display(daily_revenue.head())
```

## Plot total revenue

### Subtask:
Plot the total revenue over time.


**Reasoning**:
Plot the total revenue over time using the daily_revenue dataframe.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(daily_revenue['Data da Compra'], daily_revenue['Total Revenue'])
plt.title('Total Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.show()
```

## Calculate total revenue per store

### Subtask:
Calculate the total revenue for each store.

**Reasoning**:
Group the combined dataframe by store, sum the total revenue for each store, and display the result.

```python
# Although the dataframe `df_combined` contains data from different stores,
# there is no column to identify which row belongs to which store.
# I'll add a column to `df_combined` to identify the store for each row.
# I'll assume that the order of concatenation in `df_list` (loja, loja2, loja3, loja4)
# corresponds to the order of rows in `df_combined`.
# This is a simplification, and in a real-world scenario, the data should ideally
# have a store identifier column.

store_names = ['loja1', 'loja2', 'loja3', 'loja4']
df_combined['Store'] = ''

current_index = 0
for i, df in enumerate(df_list):
    df_combined.loc[current_index:current_index + len(df) - 1, 'Store'] = store_names[i]
    current_index += len(df)


store_revenue = df_combined.groupby('Store')['Total Revenue'].sum().reset_index()
display(store_revenue)
```

## Plot total revenue per store

### Subtask:
Plot the total revenue for each store.

**Reasoning**:
Plot a bar chart to visualize the total revenue for each store.

```python
plt.figure(figsize=(10, 6))
plt.bar(store_revenue['Store'], store_revenue['Total Revenue'])
plt.title('Total Revenue per Store')
plt.xlabel('Store')
plt.ylabel('Total Revenue')
plt.show()
```
